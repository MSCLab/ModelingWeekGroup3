```python
LON, LAT = np.meshgrid(lon, lat)
```

```python
import matplotlib.pyplot as plt

plt.figure()
qvp = plt.contourf(LON, LAT, qvv)
plt.colorbar(qvp)
plt.title('Contour of Specific Humidity')
plt.xlabel('Lon (degrees/east)')
plt.ylabel('Lat (degress/north)')
plt.show()
```

## illustrative example to use matplotlib

```python
import matplotlib.pyplot as plt

xlist = np.linspace(-3.0, 3.0, 100)
ylist = np.linspace(-2.0, 2.0, 50)
X, Y = np.meshgrid(xlist, ylist)
Z = np.sqrt(X**2 + Y**2)

plt.figure()

cp = plt.contourf(X, Y, Z)
plt.colorbar(cp)

plt.title('Filled Contours Plot')
plt.xlabel('x (cm)')
plt.ylabel('y (cm)')
plt.show()
```
